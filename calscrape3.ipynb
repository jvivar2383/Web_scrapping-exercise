```python
url = "https://www.ccny.cuny.edu/registrar/fall"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
col_tags = soup.find_all('th')
for i in col_tags:
    print(col_tags)
```

```python
data_tags = soup.find_all('td')
print(len(data_tags))
```

```python
index = []
day = []
message = []
```

```python

for i in data_tags:
    style = i.get('style')
    
    if style =="width:305px":
        
        val = i.get_text()
        
        val += ' 2021'
        
        
        index.append(val)
        
    if style == "width:327px":
        day.append(i.get_text())
        
    if style =="width:617px":
        message.append(i.get_text())
    #print("this is the tag: ",i.get_text(), style)
```

```python
k = 0
for d in day:
    d = d.replace('\n', "")
    day[k] = d
    k+=1
print(day)
```

```python
k = 0
for msg in message:
    msg = msg.replace('\n', "")
    msg = msg.replace('\t', '')
    message[k] = msg
    k+=1
print(message)
```

```python
problem_dates= []
dates = []

for string in index:
    #print(i)
    string = string.replace('\n', "")
    
    
    if '-' in string:
        split = string.split('-')
        dates.append(split[0]+ '2021')
        
    else:
        dates.append(string)


print("good dates: ", dates, "bad ones: ", problem_dates)
```

```python
date_index = []

for date in dates:
    date_obj = datetime.strptime(date, '%B %d %Y')
    date_index.append(date_obj)
    
print(date_index)
```

```python
dict_data = {'Day':day, 'Message': message }
#index.as_type(datetime)
```

```python
df = pd.DataFrame(dict_data, index = date_index)
print(df)
```
